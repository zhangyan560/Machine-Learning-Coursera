```python
def computeCost(X, Y, theta):
    m = len(Y)
    J = 1/(2*m)*(X * theta - Y).T*(X * theta - Y)
    return J
```

```python
def gradientDescent(X ,Y, theta,alpha = 0.01,num = 1000):
    m = len(Y)
    J = []
    for i in range(m):
        theta -= alpha/m*X.T*(X * theta - Y)
        J.append(computeCost(X, Y, theta))
        
    return theta, J
```
